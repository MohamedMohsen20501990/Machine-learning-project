```python
# el column elli hashta5al 3aleh we hatwaqa3 3aleh ha3meloh drop men el data we ha3meloh lewahdo data kalati
# 1 for columns 0 for rows and inplace to overwrtie the data
x = data.drop(['late_aircraft_delay'], axis=1, inplace=False)
y =data['late_aircraft_delay']
```

```python
#import libraries
from sklearn.model_selection import train_test_split

#splitting the data
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=44, shuffle=True)

# show the splitted data
print('x train shape is ', x_train.shape)
print('y train shape is ', y_train.shape)
print('x test shape is ', x_test.shape)
print('y test shape is ', y_test.shape)
```

```python
x_train.head()
```

```python
# import libraries to train the model after determining the x,y
from sklearn.linear_model import LinearRegression
```

```python
#applting the linear regression model to the objects that i have made
LinearRegressionModel = LinearRegression(fit_intercept=True, copy_X=True, n_jobs=-1)
```

```python
%%time

# the most important step, i will always use the LinearReressionModel object to train the dataa

LinearRegressionModel.fit(x_train, y_train)
```

```python
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(x_train, y_train))
```

```python


#Calculating Details
# we ageb values lelweights we el b's f(x) = wx+b ahof howa gably afdal qeyam leil w's, b
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(x_test, y_test))
print('Linear Regression Coef is : ' , LinearRegressionModel.coef_)
print('Linear Regression intercept is : ' , LinearRegressionModel.intercept_)
#print('----------------------------------------------------')
```

```python
# calculating the prediction

y_pred = LinearRegressionModel.predict(x_test)
y_pred
```
